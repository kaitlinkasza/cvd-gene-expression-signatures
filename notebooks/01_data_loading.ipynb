# Data Loading

This notebook downloads and preprocesses public
transcriptomic datasets associated with cardiovascular disease.

Dataset:
- GSE12288
- Whole blood gene expression profiles
- Coronary artery disease patients vs controls

```python
import GEOparse

gse = GEOparse.get_GEO("GSE12288", destdir="../data/raw")
```

```python
samples = []

for gsm_name, gsm in gse.gsms.items():
    temp = gsm.table[["ID_REF", "VALUE"]]
    temp.columns = ["gene", gsm_name]
    samples.append(temp)

expression = samples[0]

for sample in samples[1:]:
    expression = expression.merge(sample, on="gene")

expression.to_csv(
    "../data/processed/expression_matrix.csv",
    index=False
)
```
